```python
import matplotlib.pyplot as plt
import statistics
import random
import numpy
```

```python
#QUESTION 1
#parameters
npr = 50
lpr = 75
maxlo = 5
loprice = 15
avgd = 50
orderq = 55
```

```python
trials = 10000
sample = list()
for i in range(trials):
    demand = numpy.random.poisson(avgd)
    if demand < orderq:
        qtys = demand
    else:
        qtys = orderq
    qtyl = orderq - qtys
    salesrev = lpr * qtys
    if (orderq - qtys) >= maxlo:
        lorev = maxlo * loprice
    else:
        lorev = (orderq - qtys) * loprice
    totalcost = npr * orderq
    profit = salesrev + lorev - totalcost
    sample.append(profit)
```

```python
plt.hist(sample)
print("The histogram represents the possible profits for Lucy's bookstore based on the simulation model of 10,000 trials. The histogram shows how often or how likely each profit value occurred in the simulation. The histogram also shows that the profit distribution is negatively skewed or right modal, which means that most of the profit values are on the right side of the histogram, but there are some low profit values on the left side that pull the mean down. This indicates that Lucy has a high chance of making a good profit, but also a small risk of losing money or breaking even.")
```

```python
#Summary Statistics
print('max profit = ', max(sample))
print('median profit = ', statistics.median(sample))
print('mean profit = ', statistics.mean(sample))
print('min profit =', min(sample))
```

```python
count = 0
for j in sample:
    if j < 0:
        count = count +1
losschance = count/len(sample)
print('there is a %5.2f%% chance of a net loss' % (losschance *100))
```

```python
orderl = list(range(35, 65))
avgprofl = list()

for orderq in orderl:
    trials = 10000
    sample = list()
    for i in range(trials):
        demand = numpy.random.poisson(avgd)
        if demand < orderq:
            qtys = demand
        else:
            qtys = orderq
        qtyl = orderq - qtys
        salesrev = lpr * qtys
        if (orderq - qtys) >= maxlo:
            salvrev = maxlo * loprice
        else:
            salvrev = (orderq - qtys) * loprice
        totalcost = npr * orderq
        profit = salesrev + salvrev - totalcost
        sample.append(profit)
    samplemean = statistics.mean(sample)
    avgprofl.append(samplemean)
```

```python
#QUESTION 2
#parameters
ppumin = 18.95
ppumax = 26.95
ppumode = 24.95
costpumin = 12
costpumax = 15
qsint = 10000
qsx = -250
fcavg = 30000
fcsd = 5000
rvm = 0
rvsd = 10
```

```python
trials = 10000
cef = list()
for i in range(trials):
    sppu = random.triangular(ppumin, ppumax, ppumode)
    cpu = random.uniform(costpumin, costpumax)
    qs = qsint + qsx * sppu + random.normalvariate(rvm, rvsd)
    fc = random.normalvariate(fcavg, fcsd)
    tc = fc + cpu * qs
    trev = qs * sppu
    profto = trev - tc
    cef.append(profto)
```

```python
plt.hist(cef)
print('The histogram below represents the distribution of profits for the consumer electronics firm based on the simulation model of 10,000 trials above. This shows how often or how likely each profit value occured in the simulation. Additionally, the shape resembles a bell-curve, although not a perfect one. This means that the mean, median, and mode are all very close to one another')
```

```python
print('max profit =  %7.2f' % max(cef))
print('median profit = %7.2f' % statistics.median(cef))
print('mean profit = %7.2f' % statistics.mean(cef))
print('min profit = %7.2f' % min(cef))
```
