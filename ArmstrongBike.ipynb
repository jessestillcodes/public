```python
# ABC Flyer Frame Model Parameters
unit_cost = 100.0
intercept = 750
slope = -5.0
```

```python
#Decision Variable
price = 160.00
```

```python
# Model Calculations
demand = intercept + slope  * price
if demand < 0:
    demand = 0
cost = unit_cost * demand
revenue = price * demand
profit = revenue - cost

print('price = ', price, 'revenue = ', revenue, 'cost = ', cost, 'profit = ', profit)
```

```python
price_list = list(range(100,165,5))
print(price_list)
```

```python
profit_list = list()
for price in price_list:
    demand = intercept + slope * price
    if demand < 0:
        demand = 0
    cost = unit_cost * demand
    revenue = price * demand
    profit = revenue - cost
    print('price = ', price, 'profit = ', profit)
    profit_list.append(profit)
```
